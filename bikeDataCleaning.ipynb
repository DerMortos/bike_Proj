```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_columns', 500)
bike_df = pd.read_csv('./files/bike_sharing_dataset.csv')
bike_df.head()
```

```python
# initial data peek
bike_df.describe()
```

```python
bike_df.dtypes
```

```python
bike_df.isnull().sum()
```

```python
# replacing NaN with 0 for weather and holiday data
wtFeatures = [x for x in bike_df.columns if 'wt' in x]
bike_df['holiday'] = bike_df['holiday'].fillna(0)
bike_df[wtFeatures] = bike_df[wtFeatures].fillna(0)
```

```python
missingTarget = bike_df[bike_df['total_cust'].isna()]
missingTarget
```

```python
#using forard fill for missing customers in time series data
bike_df[['total_cust', 'casual', 'registered']] = bike_df[['total_cust', 'casual', 'registered']].fillna(method='ffill')
bike_df.isnull().sum()
```
